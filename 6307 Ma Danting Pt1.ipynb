```python
#Задание 3
#Даны два одномерных нормальных распределения Na и Nb с мат. ожиданиями 4, 8 и СКО 1, 2 соответственно.  
Na_mean =4  
Nb_mean =8  
Na_std = 1  
Nb_std = 2  
Na = np.random.normal(Na_mean, Na_std, 1000)  
plt.figure(figsize=(20,10),dpi=100)  
plt.hist(Na,25)  
plt.show() 
```

```python
Nb = np.random.normal(Nb_mean, Nb_std, 1000)  
plt.figure(figsize=(20,10),dpi=100)  
plt.hist(Nb,25)  
plt.show()
```

```python
#Для каждого из значения {5,6,7} определите какое из распределений сгенерировало значение с большей вероятностью.  
data = [5,6,7]  
Pa = stats.norm.pdf(data, Na_mean, Na_std)  
Pb = stats.norm.pdf(data, Nb_mean, Nb_std)  
['Na' if difP > 0 else 'Nb' for difP in Pa - Pb]
```

```python
#Найди значение, которой могло быть сгенерировано обеими распределениями с равной вероятностью  
x = np.linspace(5, 6, 100)  
Pa = stats.norm.pdf(x, Na_mean, Na_std)  
Pb = stats.norm.pdf(x, Nb_mean, Nb_std)  
result = [x[i] for i in range(len(Pa)) if abs(Pa[i] - Pb[i]) <= 0.001]  
result
```
